<h1 style="font-size: 40px;">2. DATA SCRAPING </h1>

```python
# URL of the webpage to scrape
url = "https://fbref.com/en/comps/Big5/stats/players/Big-5-European-Leagues-Stats"

# Send a GET request to the URL and get the HTML content
response = requests.get(url).text
```

```python
from bs4 import BeautifulSoup
```

```python
# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response)
```

```python
table = soup.find_all('table')
```

```python
ta = table[0]
html_string = str(ta)
df = pd.read_html(io.StringIO(html_string))[0]
```

Reformat the columns "Nation" and "Comp". Meanwhile, remove the column "Born", which has the same function to "Age"

```python
df['Nation'] = df['Nation'].str.split(r' ').str.get(1)
```

```python
df['Comp'] = df['Comp'].str.split(r' ').str.get(1)
```

```python
df['Comp'] = df['Comp'].str.replace('Premier','Premier League')
df['Comp'] = df['Comp'].str.replace('Ligue','Ligue 1')
df['Comp'] = df['Comp'].str.replace('Serie','Serie A')
df['Comp'] = df['Comp'].str.replace('La','La Liga')
```

```python
df = df.drop(['Born'],axis=1)
df
```

```python
df = df.drop(df[df['Player'] == 'Player'].index)
```
