```python
df = pd.read_csv('Defencee.csv')
```

```python
df.index=df['Rk']
df = df.drop(df.columns[:3], axis=1)
```

```python
df['Nation'] = df['Nation'].str.split(r' ').str.get(1)
df['Comp'] = df['Comp'].str.split(r' ').str.get(1)
df['Comp'] = df['Comp'].str.replace('Premier','Premier League')
df['Comp'] = df['Comp'].str.replace('Ligue','Ligue 1')
df['Comp'] = df['Comp'].str.replace('Serie','Serie A')
df['Comp'] = df['Comp'].str.replace('La','La Liga')
```

```python
df = df.drop(['Born'],axis=1)
df
```

```python
df = df.drop(df[df['Player'] == 'Player'].index)
```

```python
df2 = df.copy()
```

```python
df2.drop(df2.columns[-1],inplace=True,axis=1)
```

```python
df2.drop(df2.columns[:6],inplace=True,axis=1)
```

```python
df2.drop(df2.columns[3:6],inplace=True,axis=1)
```

```python
df2.drop(df2.columns[6],inplace=True,axis=1)
```

```python
df2.drop(df2.columns[10],inplace=True,axis=1)
```

```python
df3 = df2.copy()
df3 = df3.rename(columns={'Tkl': 'TklAPlayers90','TklW': 'TklW90','Tkl.1':'TklADribblers90','Att':'TklAttDri90','Blocks':'Bl90','Sh':'ShBl90','Pass':'PassBl90','Int':'Int90','Clr':'Clr90','Err':'Err90'})
df3
```

```python
df3 = df3.astype(float)
```

```python
df3['TklAPlayers90']=df3['TklAPlayers90']/df3['90s']
df3['TklW90']=df3['TklW90']/df3['90s']
df3['TklADribblers90']=df3['TklADribblers90']/df3['90s']
df3['TklAttDri90']=df3['TklAttDri90']/df3['90s']
df3['Bl90']=df3['Bl90']/df3['90s']
df3['ShBl90']=df3['ShBl90']/df3['90s']
df3['PassBl90']=df3['PassBl90']/df3['90s']
df3['Int90']=df3['Int90']/df3['90s']
df3['Clr90']=df3['Clr90']/df3['90s']
df3['Err90']=df3['Err90']/df3['90s']
```

```python
df3 = df3.round(2)
```
