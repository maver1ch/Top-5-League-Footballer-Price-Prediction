```python
df = pd.read_csv('Possessionn.csv')
```

```python
df.index=df['Rk']
df = df.drop(df.columns[:3], axis=1)
```

```python
df['Nation'] = df['Nation'].str.split(r' ').str.get(1)
df['Comp'] = df['Comp'].str.split(r' ').str.get(1)
df['Comp'] = df['Comp'].str.replace('Premier','Premier League')
df['Comp'] = df['Comp'].str.replace('Ligue','Ligue 1')
df['Comp'] = df['Comp'].str.replace('Serie','Serie A')
df['Comp'] = df['Comp'].str.replace('La','La Liga')
```

```python
df = df.drop(['Born'],axis=1)
df
```

```python
df = df.drop(df[df['Player'] == 'Player'].index)
```

```python
df2 = df.copy()
```

```python
df2
```

```python
df2.drop(df2.columns[:6],inplace=True,axis=1)
```

```python
df2.drop(df2.columns[2:8],inplace=True,axis=1)
```

```python
df2.drop(df2.columns[5:7],inplace=True,axis=1)
```

```python
df2 = df2.astype(float)
```

```python
df3 = df2.copy()
```

```python
df3 = df3.rename(columns={'Touches': 'Touches90','Att': 'TakeOnAtt90','Succ':'TakeOnSucc90','Carries':'Carries90','TotDist':'CarriesTotDist90','PrgDist':'CarriesPrgDist90','PrgC':'PrgC90','1/3':'Caries1/3(90)','PrgR':'PrgR90','CPA':'CPA90','Mis':'Mis90','Dis':'Dis90','Rec':'Rec90'})
```

```python
df3
```

```python
df3['Touches90']=df3['Touches90']/df3['90s']
df3['TakeOnAtt90']=df3['TakeOnAtt90']/df3['90s']
df3['TakeOnSucc90']=df3['TakeOnSucc90']/df3['90s']
df3['Carries90']=df3['Carries90']/df3['90s']
df3['CarriesTotDist90']=df3['CarriesTotDist90']/df3['90s']
df3['CarriesPrgDist90']=df3['CarriesPrgDist90']/df3['90s']
df3['PrgC90']=df3['PrgC90']/df3['90s']
df3['Caries1/3(90)']=df3['Caries1/3(90)']/df3['90s']
df3['CPA90']=df3['CPA90']/df3['90s']
df3['Mis90']=df3['Mis90']/df3['90s']
df3['Dis90']=df3['Dis90']/df3['90s']
df3['Rec90']=df3['Rec90']/df3['90s']
df3['PrgR90']=df3['PrgR90']/df3['90s']
df3
```
