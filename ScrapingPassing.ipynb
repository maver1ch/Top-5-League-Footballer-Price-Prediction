```python
df = pd.read_csv('Passingg.csv')
```

```python
df.index=df['Rk']
df = df.drop(df.columns[:3], axis=1)
```

```python
df['Nation'] = df['Nation'].str.split(r' ').str.get(1)
df['Comp'] = df['Comp'].str.split(r' ').str.get(1)
df['Comp'] = df['Comp'].str.replace('Premier','Premier League')
df['Comp'] = df['Comp'].str.replace('Ligue','Ligue 1')
df['Comp'] = df['Comp'].str.replace('Serie','Serie A')
df['Comp'] = df['Comp'].str.replace('La','La Liga')
```

```python
df = df.drop(['Born'],axis=1)
df
```

```python
df = df.drop(df[df['Player'] == 'Player'].index)
```

```python
df2 = df.copy()
```

```python
df2.drop(df2.columns[-1],inplace=True,axis=1)
```

```python
df2.drop(df2.columns[6:16],inplace=True,axis=1)
```

```python
df3 = df2.copy()
df3 = df3.rename(columns={'Cmp': 'Cmp90','Att': 'Att90','TotDist':'PassTotDist90','PrgDist':'PassPrgDist90','KP':'KP90','1/3':'Pass1/3(90)','PPA':'PPA90','CrsPA':'CrsPA90','PrgP':'PrgP90'})
df3
```

```python
df3.dtypes
```

```python
df3 = df3.astype(float)
```

```python
df3['Cmp90']=df3['Cmp90']/df3['90s']
```

```python
df3['Att90']=df3['Att90']/df3['90s']
df3['PassTotDist90']=df3['PassTotDist90']/df3['90s']
df3['PassPrgDist90']=df3['PassPrgDist90']/df3['90s']
df3['KP90']=df3['KP90']/df3['90s']
df3['Pass1/3(90)']=df3['Pass1/3(90)']/df3['90s']
df3['PPA90']=df3['PPA90']/df3['90s']
df3['CrsPA90']=df3['CrsPA90']/df3['90s']
df3['PrgP90']=df3['PrgP90']/df3['90s']
```

```python
df3 = df3.round(2)
```

```python
df3.drop(df2.columns[0],inplace=True,axis=1)
```
