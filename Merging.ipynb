```python
df = pd.merge(df_1, df_2, on="Rk")
df = pd.merge(df, df_3, on="Rk")
df = pd.merge(df, df_4, on="Rk")
df = pd.merge(df, df_5, on="Rk")
```

```python
df
```

### Preprocessing

1. Remove all the goalkeepers whose specificity interfere with the prediction process. Besides, data on goalkeeper characteristics are also limited.
2. Remove all the players who played less than 600 minutes because of unproven performance and low reliability of the data.
3. There are 18 pairs of players with identical names, so removing them does not significantly influence on the overall data.

```python
df = df[df['Pos'] != 'GK']
```

```python
df = df[df['Min'] >= 600]
```

```python
duplicates = df[df.duplicated('Player', keep=False)]['Player']
print(duplicates)
```

```python
df = df.drop_duplicates(subset=['Player'])
```
